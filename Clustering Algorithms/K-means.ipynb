```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=3)
```

```python
X #X is a numpy array containing the x and y coordinates of the points
```

```python
y #y is also a numpy array that contains the values from 0-5(no. of blobs) to denote which point belongs to which group/blob
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

### we have create 6 groups of datapoints successfully

```python
k=5
colors=["red", "green", "blue", "yellow", "purple"] 
clusters = {}
for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1) #randomly choosing a point as the centre
    point = []
    cluster = { "center":center, "points":point, "color":colors[i]}
    clusters[i] = cluster
```

```python
clusters
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
for i in range(k):
    center = clusters[i]["center"]
    plt.scatter(center[0], center[1], c="black", marker="^")
```

```python
def distance(v1, v2):
    dist = np.sqrt(np.sum((v1-v2)**2))
    return dist
```

```python
v1 = np.array([1,2])
v2 = np.array([3,4])
print(distance(v1, v2))
```

```python
def assign_cluster():
    for idx in range(X.shape[0]):
        dist = []
        curr_x = X[idx]
        for i in range(k):
            d = distance(curr_x, clusters[i]["center"])
            dist.append(d)
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]["points"].append(curr_x)
```

```python
def update_cluster():
    for idx in range(k):
        pts = np.array(clusters[idx]["points"])
        if pts.shape[0]>0:
            new_center = pts.mean(axis=0)
            clusters[idx]["center"] = new_center
            clusters[idx]["points"] = []
```

```python
def plot_cluster():
    for i in clusters:
        pts = np.array(clusters[i]["points"])
        try:
            plt.scatter(pts[:, 0], pts[:, 1], c=clusters[i]["color"])
        except:
            pass
        center = clusters[i]["center"]
        plt.scatter(center[0], center[1], c="black", marker="^")
```

```python
assign_cluster()
plot_cluster()
update_cluster()
```

## Scikit Learn K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 5)
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=3)
```

```python
km.fit(X, y)
```

```python
centers = km.cluster_centers_
```

```python
centers
```

```python
labels = km.labels_
```

```python
labels
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
for i in range(k):
    center = centers[i]
    plt.scatter(center[0], center[1], marker="^", c="black")
```
